```python
import flatspin.model as md
import matplotlib.pyplot as plt
import numpy as np
```

```python
model = md.PinwheelSpinIceLuckyKnot(size=(25,25), alpha=0.1, init='random', use_opencl=True)

E=model.energy()
E_tot=model.total_energy()
print("Energy (randomized):", E)
print("sum(Energy) (randomized):", np.sum(E))
print("Total energy (randomized):", E_tot)

plt.figure()
quiv = model.plot(C=E, cmap='plasma')
plt.colorbar(quiv);
```

```python
model.relax()
#model.plot_vertex_mag();

E=model.energy()
E_tot=model.total_energy()
print("Energy (randomized):", E)
print("sum(Energy) (randomized):", np.sum(E))
print("Total energy (randomized):", E_tot)

plt.figure()
quiv = model.plot(C=E, cmap='plasma')
plt.colorbar(quiv);
```

```python
model = md.PinwheelSpinIceLuckyKnot(size=(25,25), alpha=0.1, init='random', use_opencl=True)
model.relax()

# Saturating field
model.set_h_ext([0.2, 0.2])

E = model.energy()
print("Total energy (with h_ext):", np.sum(E))
print("Total energy (with h_ext):", model.total_energy())

quiv = model.plot(C=E, cmap='plasma')
plt.colorbar(quiv);
```

```python
model = md.PinwheelSpinIceLuckyKnot(size=(25,25), alpha=0.1, init='random', use_opencl=True)
model.relax()

# Saturating field
model.set_h_ext([0.2, 0.2])

E = model.energy()
model.relax()
E = model.energy()
print("Total energy (with h_ext):", np.sum(E))
print("Total energy (with h_ext):", model.total_energy())

quiv = model.plot(C=E, cmap='plasma')
plt.colorbar(quiv);
```
